# Nuevo target: dieléctrico de espesor D sobre conductor perfecto.

El objetivo será encontrar la la constante dieléctrica que mejor caracteriza a la placa dieléctrica. Para ello vamos a realizar las siguientes mediciones:

- S0: medición sin target: rebote de la señal sobre el piso de la habitación $\rightarrow$ *background*.
- S1: medición con placa conductora.
- S2: medición con dieléctrico sobre placa conductora.

* **S_cal = S1 - S0**: señal de calibración.
* **S_target = S2 - S0**: señal reflejada por el target de interés.

> **La señal de calibración** es la que, idealmente, emite la antena. Si uno pudiese modelar de punta a punta la antena y hacer todas las cuentas necesarias, obtendría que la antena emite el pulso $S_{cal}(t)$, o bien su transformada de Fourier $\hat{S}_{cal}(f)$. Esto lo decimos porque medimos esta señal luego de la reflexión con un target que tiene el coeficiente de reflexión $|R|=1$. Está claro que esto es una hipótesis fuerte, que deberá ser corregida luego. Además, falta contemplar el factor de decaimiento de la señal debido al camino de ida-vuelta entre la antena emisora y la antena receptora. Pero este factor -geométrico- es independiente de las frecuencias del paquete de ondas y del target, así que por ahora podemos ignorarlo.


> **La señal del target** es la que debemos analizar para obtener alguna de las propiedades del blanco en estudio. En principio desconocemos la constante dieléctrica $\epsilon(\omega)$ que tiene la placa dieléctrica de espesor $D$ puesta sobre el conductor ideal. Y, en principio, la señal reflejada por este target que recibe la antena  es $S_{teo}=R(\omega)\,\times\,S_{antena}(\omega)=R(\omega)\,\times\,S_{cal}(\omega)$. Entonces, aquí tenemos que comparar $S_{teo}$ vs $S_{target}$ y ver qué tan bien andamos rumbeados. Para ello debríamos saber de antemano la constante dieléctrica del target, que sería el parámetro a obtener! Pero bueno, este es el experimento de calibración del método UWB. Así que en este caso asumimos un valor (o varios) para $\epsilon(\omega)$ y completamos el ciclo de análisis de datos.

### Función para pasar la señal medida de banda pasante a banda base ###

```python
def to_baseband(x,Ts,BW,fc):
  
  '''
  Señal de banda pasante a banda base:
  
      Entradas:
      - x: señal de banda pasante---> primero hacer promedio de las 10 mediciones del GUI;
      - Ts: periodo de muestreo de señal;
      - BW: ancho de banda filtro;
      - fc: frec portadora.
      
      Salida:
      - señal en banda base.
  '''
  
  filtorder = 80
  cutoff = BW/2
  fs = 1/Ts
  #    h,a = signal.ellip(numtaps-1, rp=1, rs=60, Wn=cutoff, btype='lowpass', analog=False, output='ba', fs=fs)
  h = signal.firwin(filtorder, cutoff, width=None, window='hamming', pass_zero='lowpass', scale=True, fs=fs)
  time = np.arange(x.size)*Ts
  y = signal.lfilter(b=h,a=1,x = 2*x*np.exp(-1j*2*np.pi*fc*time))
  
  return y

def to_dB(x,norm=None):
  # Convierte a dB la señal x. Si se pasa un segundo argumento, normaliza a 0 dB la máxima amplitud
  x_abs = np.abs(x)
  x_abs_max = np.amax(x_abs)
  if norm:
    x_dB = 10*np.log10(x_abs/x_abs_max)
  else:
    x_dB = 10*np.log10(x_abs)
  return x_dB
```

### Función que calcula FFT de una señal $S(t)$ ###

```python
def FFT_ordenada(y,n,dt):

  '''
  Transformo Fourier la señal y(x).
      
      Entradas:
      - y: señal a transformar;
      - n: cantidad de datos de la señal;
      - dt: tiempo de muestreo.
      
      Salidas:
      - vector Xf: frecuencias de la FFT ordenadas de la más negativa a la más positiva; 
      - vector Yf: espectro de la señal transformada.
  '''
    
  xf = fftfreq(n,dt) # frecuencias de la FFT ordenadas así: primero f>o, luego f<0
  Xf = fftshift(xf) # frecuencias reordenadas
  
  Y = fftshift(fft(y,n)) # FFT de la señal, con el espectro reordenado. Comentario de Ovi: Agregué fft(), ya que fftshift sólo reacomoda la señal
    
  return Xf,Y # modifique esto para que pase los arrays separados, ya que son señales distintas.
```

### Coeficiente de reflexión a incidencia normal: target de espesor *d* y constante dieléctrica *ep* y conductor ideal debajo del dieléctrico.


```python
def Ref_1capa(ep,d,f): # antes era Ref2. Solo valido para frecuencias positivas.

  ki = 2*np.pi*f/3e8  
  fi = 2*1j*np.sqrt(ep)*ki*d

  a = 1-np.sqrt(ep)+(1+np.sqrt(ep))*np.exp(fi)
  b = 1+np.sqrt(ep)+(1-np.sqrt(ep))*np.exp(fi)

  return a/b
```

Coeficiente de reflexión a incidencia normal: target de espesor $d_1 + d_2 = d$, donde la constante dieléctrica varía en dos tramos: $\epsilon_1$ arriba, $\epsilon_2$ debajo, luego el conductor ideal. Sugerencia: $\epsilon_1 > \epsilon_2$, porque el estrato superior está más húmedo que el inferior.

```python
def Ref_2capas(ep1,ep2,d1,d2,f):

  k0 = 2*np.pi*f/3e8
  k1 = np.sqrt(ep1)*k0
  k2 = np.sqrt(ep2)*k0

  z1 = np.exp(-2*1j*k1*d1)
  z2 = np.exp(-2*1j*k2*d2)

  r01 = (np.sqrt(ep1)-1)/(np.sqrt(ep1)+1) # interface ep0-ep1; ep0 = 1
  r12 = (np.sqrt(ep2)-np.sqrt(ep1))/(np.sqrt(ep2)+np.sqrt(ep1))  # interface ep1-ep2, dentro del dieléctrico
  r2cond = 1 # interface ep2-conductor ideal
  
  a = r01 + r12/z1 + r01*r12*r2cond/z2 + r2cond/(z1*z2)
  b = 1 + r01*r12/z1 + r12*r2cond/z2 + r01*r2cond/(z1*z2)

  R = a/b

  return R
```

### Coeficiente de reflexión a incidencia normal: target de espesor $d$ ,constante dieléctrica $\epsilon_1$ y conductor real con conductividad $\sigma(\omega)$.

```python
def R_ep_cond(ep1,d,sigma,f):

  w = 2*np.pi*f
  ki = w/3e8
  k1 = np.sqrt(ep1)*ki

  r01 = (np.sqrt(ep1)-1)/(np.sqrt(ep1)+1)
  #ep_c = 1 + 1j*4*np.pi*sigma/w 
  #r1c = (np.sqrt(ep_c)-np.sqrt(ep1))/(np.sqrt(ep_c)+np.sqrt(ep1))
  delta = 3e8/np.sqrt(2*np.pi*sigma*w)
  r1c = 1-2*w*delta/3e8

  fase = 2*1j*k1*d

  a = r01+r1c*np.exp(fase)
  b = 1+r01*r1c*np.exp(fase)
  r = a/b

  return r
```

```python
## Función que transforma humedad del suelo a cte dielectrica

def mv_to_ep(mv,S,C,mv_std):
  
  ''' Convertir mv a ep usando el modelo de Hallikainen (1985)
  
  Entradas:
  - S: fracción de arena
  - C: fracción de arcilla
  - mv: humedad del suelo
  - Propaga errores en caso de que ingrese mv_std

  Salidas:
  - ep = (a0 + a1*S + a2*C) + (b0 + b1*S + b2*C)*mv + (c0 + c1*S + c2*C)*mv**2
  - Delta ep = A + (B-1)*ep
  '''

  # Coeficientes calculados a 8GHz

  # Coeficientes para la parte real
  a0r, a1r, a2r = 1.997, 0.002, 0.018
  b0r, b1r, b2r = 25.579, -0.017, -0.412
  c0r, c1r, c2r = 39.793, 0.723, 0.941

  # Coeficientes para la parte imag
  a0i, a1i, a2i = -0.201, 0.003, 0.003
  b0i, b1i, b2i = 11.266, -0.085, -0.155
  c0i, c1i, c2i= 0.194, 0.584, 0.581

  ar = a0r+a1r*S+a2r*C
  br = b0r+b1r*S+b2r*C
  cr = c0r+c1r*S+c2r*C

  ai = a0i+a1i*S+a2i*C
  bi = b0i+b1i*S+b2i*C
  ci = c0i+c1i*S+c2i*C

  ep_real = ar + br*mv + cr*mv**2
  ep_imag = ai + bi*mv + ci*mv**2

  ep = ep_real+1j*ep_imag

  if(mv_std!=0):

    deps_realdmv = br + 2*cr*mv
    deps_imagdmv = bi + 2*ci*mv
    ep_real_std = abs(deps_realdmv)*mv_std
    ep_imag_std = abs(deps_imagdmv)*mv_std 
    ep_error = ep_real_std+1j*ep_imag_std

  else:
    ep_error = 0

  return {'ep': ep, 'ep_error': ep_error}
```

## Ajuste con el modelo que incluye rugosidad

```python
MV_min = 0.095 # min
MV_max = 0.35 # med

MV=MV_min

Thi = 1e-5
D = 0.09
d1 = 0.205*D 
d2 = D-d1
Sand = 0.12
Clay = 0.17

Ep =  mv_to_ep(MV,Sand,Clay,0)['ep']
EP1 = Ep
MV2 = 0.5*MV
EP2 = mv_to_ep(MV2,Sand,Clay,0)['ep']
EP_cond = 5e8

S1 = 4.2e-3
L1 = 0.5
S2 = 1e-5
L2 = 1

Dist = 0.09

f_pos =frec_fft[frec_fft>=0]
f_pos[0] = 1e-4
n_frec = len(f_pos)
Ki_pos = 2*np.pi*f_pos/3e8
R0 = a0HH(Ki_pos,Thi,EP0,EP1,EP_cond,D) # a0 evaluado en f>=0
R2 = np.zeros(n_frec,dtype=complex)


n_gauss = 75
for j in range(n_frec):

    frec = f_pos[j]
    Ki = Ki_pos[j]

    k_lim = 2.5*Ki

    RX = IG.int_gauss(n_gauss,k_lim)['X_IG']
    RY = IG.int_gauss(n_gauss,k_lim)['Y_IG']
    Wt = IG.int_gauss(n_gauss,k_lim)['Wt']
    m_gauss = IG.int_gauss(n_gauss,k_lim)['m_gauss']
    J = k_lim**2
    

    aux = R_rugoso(RX,RY,EP1,S1,L1,S2,L2,Dist,frec)
    R2[j] = J*np.nansum(np.nansum(Wt*np.reshape(aux,(1,m_gauss**2))))


S_teo_spm_fft = S_cal_fft[512:1024]*(R0+R2)


# S_teo_fft = S_cal_fft
plt.figure(figsize=(21,8))
ind = np.array([0,1,2,3,4,5,6,7])
#for i in range(len(ind)):
 # plt.plot(frec_fft/1e9,S_tar_fft[:,ind[i]],'--',label='Mediciones:  '+str(humedades[ind[i]]/10))
plt.plot(f_pos/1e9,np.abs(S_tar_fft[512:1024,ind[0]]),'--',label='Mediciones:  '+str(humedades[ind[0]]/10))
plt.plot(f_pos/1e9,np.abs(S_teo_spm_fft),'r--',label='Estimado SPM')
plt.plot(f_pos/1e9,np.abs(S_teo_1ep_fft[512:1024]),'b*--',label='Estimado Fresnel')
plt.xlabel('Frec. [GHz]')
plt.ylabel('R')
plt.grid('on')
plt.legend(loc="best")
plt.title('Señales Medidas vs Señales Estimadas')
plt.xlim([6,10])
# plt.ylim([-35,0])
plt.show()
```

```python
## ------------------ ajuste de las mediciones --------------------- ##

# grilla de valores para MV
mv_min, mv_max, n_mv = 0.75*np.min(mv_med),1.05*np.max(mv_med), 250
MV = np.linspace(mv_min,mv_max,n_mv)

D = 9e-2  # dist en cm

#parametros del modelo de tierra
Sand = 0.12
Clay = 0.17

ep_std = 0.015  # std para estimar el error en ep del modelo teórico
EP1 =  mv_to_ep(MV,Sand,Clay,ep_std)['ep']
EP1_error = mv_to_ep(MV,Sand,Clay,ep_std)['ep_error']


## Me quedo con la parte de f>=0 de las señales medidas
f_min = 6.5e9  # frec de corte mínima
f_max = 10.5e9 # frec de corte máxima
f_pos = frec_fft[(frec_fft>f_min) & (frec_fft<f_max)]
n_frecs = len(f_pos)
start = np.min(np.where(frec_fft>f_min))
end = int(np.max(np.where(frec_fft<f_max))+1)
f_norm = f_pos/1e9 # frec >0 normalizada en GHz

S_cal_fpos = S_cal_fft[start:end]     #Background
S_tar_fpos = S_tar_fft[start:end,:]   #S2 medida


## inicio el ajuste para cada medición (índice i0)
for i0 in range(n_ep):                #para cada paso del vector de epsilons
  for i1 in range(n_mv):              #para cada medicion
    ep1 = EP1[i1]                     #defino el epsilon del paso
    R_fpos = Ref_1capa(ep1,D,f_pos)   #calculo R
    S_teo_fpos = S_cal_fpos*R_fpos    #calulo S2 teorico

    ## Calculo métricas de error: error absoluto medio ##
    e_mae[i0,i1] = np.sum(np.abs(np.abs(S_teo_fpos)-np.abs(S_tar_fpos[:,i0])))/len(S_teo_fpos) 
    

  # busco el índice en donde e_mae es mínimo para la medición i0
  ind = np.argmin(e_mae[i0,:])

  # señal teórica óptima para el ajuste de la medición i0
  mv_optimo[i0] = MV[ind]                                   #humedad volumetrica
  ep_optimo[i0,0] = EP1[ind]                                #cte dieléctrica
  ep_optimo[i0,1] =  EP1_error[ind]                         #error de epsilon
  S_optima[i0,:] = S_cal_fpos*Ref_1capa(EP1[ind],D,f_pos)   #S2 teorica

  # guardo el mínimo error para cada medición
  errores_mae[i0]=e_mae[i0,ind]
```
