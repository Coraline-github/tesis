Funcion que cuenta el numero de archivos en un directorio y me da los tamaños

```python
def check_import_ub(path, portadora, altura, target):
    '''
    Ubuntu - Funcion que chequea numerode mediciones y long. Importa datos
    Inputs:
    Path = path donde estan las mediciones
    Portadora = Fc8/ o Fc7/
    altura = 1, 2, 3 o 4 segun altura de med
    target = S1 o S2

    Output:
    S = devuelve el vector S1 o S2 segun lo que importo
    '''

    # cuento el número de archivos en el directorio
    list = os.listdir(path+target+'/'+portadora+altura) # dir is your directory path
    N_med = int(len(list))
    print('# mediciones =', N_med)
    print()

    # miro la longitud de cada mediciòn
    L_data = np.zeros(N_med, dtype=int) # vector en donde guardo la long de cada medición

    for j in range(N_med): 
        J = str(j)
        len_data = int(len(open(path+target+'/'+portadora+altura+target+'_'+J+'.txt').readlines()))
        L_data[j] = len_data
        print('medición ',target,' #'+J+' =',len_data)
    
    S = np.zeros((N_med,int(len_data-1),11),dtype=float)
    
    # Python program to check if all elements in a List are same 
    first = L_data[0]
    chk = True

    # Comparing each element with first item 
    for item in L_data:
        if first != item:
            chk = False
            break;
              
    if (chk == True): 
        print("iguales, importo datos")

        for j0 in range(N_med):
            J0 = str(j0)
            S[j0,:,:] = np.genfromtxt(path+target+'/'+portadora+altura+target+'_'+J0+'.txt', delimiter=' ')

    else: 
        print("diferentes. Editar datos")      

    return S    
  
```

```python
def Z2_check_import_win(path, portadora, altura):

    '''
    Windows - Funcion que chequea numerode mediciones y long. Importa datos
    Inputs:
    Path = path donde estan las mediciones
    Portadora = Fc8/ o Fc7/
    altura = 1, 2, 3 o 4 segun altura de med
    tita = angulos

    Output:
    S = devuelve el vector S1 o S2 segun lo que importo
    '''

    #cuento el numero de archivos Z2 e el directorio
    med_Z = os.listdir(path+'\\'+'Z2'+'\\'+portadora+'\\')
    N_med_Z = int(len(med_Z))
    print()
    print('# de angulos =', N_med_Z)

    for j in range(N_med_Z): 
        J = str(j+1)

        #veo cantidad de repeticiones
        rep = os.listdir(path+'\\'+'Z2'+'\\'+portadora+'\\'+'tita'+J)
        N_rep = int(len(rep))

        print('# de repeticiones =', N_rep)

        # miro la longitud de cada mediciòn Z2
        L_data_Z2 = np.zeros(N_rep, dtype=int)

        for k in range(N_rep):
            K = str(k+1)
            len_data_Z2 = int(len(open(path+'\\'+'Z2'+'\\'+portadora+'\\'+'tita'+J+'\\'+K+'.txt').readlines()))
            L_data_Z2[k] = len_data_Z2
            print('longitud medición #'+K+' =',len_data_Z2)
        print()
    
    
    Z2 = np.zeros((int(len_data_Z2-1),11,N_med_Z,N_rep),dtype=float)
    
    # Python program to check if all elements in a List are same 
    first = L_data_Z2[0]
    chk = True

    # Comparing each element with first item 
    for item in L_data_Z2:
        if first != item:
            chk = False
            break;
              
    if (chk == True): 
        print()
        print("iguales, importo datos!!")

        for j0 in range(N_med_Z):
            J0 = str(j0+1)

            for k0 in range(N_rep):
                K0 = str(k0+1)

                Z2[:,:,j0,k0] = np.genfromtxt(path+'\\'+'Z2'+'\\'+portadora+'\\'+'tita'+J0+'\\'+K0+'.txt', delimiter=' ')

    else: 
        print("diferentes. Editar datos!!")      

    return Z2
  
```
