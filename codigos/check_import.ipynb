## Funcion que cuenta el numero de archivos en un directorio y si son todos iguales me importa los datos de la medicion

```python
import numpy as np
import os
```

```python
def check_import_ub(path, target, portadora):
    '''
    Ubuntu - Funcion que chequea numero de mediciones y long. Importa datos
    Inputs:
    Path = path donde estan las mediciones
    (ej: /home/gordotony/Documents/GitHub/tesis/mediciones_MM_DD)
    Portadora = Fc8 o Fc7
    target = (S1, S2, M2, X2) o (Z2, MZ2)

    Output:
    h = alturas de las mediciones
    S = devuelve el vector segun lo que import
    tita = angulo de incidencia (si es dif a 0)
    '''
    
    h = np.genfromtxt(path+'/'+target+'/h.txt', delimiter=' ')
    alturas = len(h)

    if target == 'Z2' or target == 'MZ2':
        tita = np.genfromtxt(path+'/'+target+'/tita.txt', delimiter=' ')
        ang = len(tita)

        #suelo medir en una unica altura asi que no tengo en cuenta la variable h

        for l in range(ang):
            L = str(l+1)
            # cuento el número de archivos en el directorio
            list = os.listdir(path+'/'+target+'/'+portadora+'/tita'+L)
            N_med = int(len(list))
            #print('# mediciones =', N_med)
            #print()

            # miro la longitud de cada mediciòn
            L_data = np.zeros(N_med, dtype=int) # vector en donde guardo la long de cada medición

            for j in range(N_med): 
                J = str(j+1)
                len_data = int(len(open(path+'/'+target+'/'+portadora+'/tita'+L+'/'+J+'.txt').readlines()))
                L_data[j] = len_data
                #print('medición ',target,' #'+J+' =',len_data)

            #-----------------------------------------------------------------
            # Python program to check if all elements in a List are same 
            #-----------------------------------------------------------------
            first = L_data[0]
            chk = True

            # Comparing each element with first item 
            for item in L_data:
                if first != item:
                    chk = False
                    break;
                    
            if (chk == True): 
                #print()
                #print("iguales, importo datos!!")
                S = np.zeros((ang,N_med,int(len_data-1),11),dtype=float)

                for j0 in range(N_med):
                    J0 = str(j0+1)
                    S[l,j0,:,:] = np.genfromtxt(path+'/'+target+'/'+portadora+'/tita'+L+'/'+J+'.txt', delimiter=' ')

            else: 
                print("diferentes. Editar datos!!")

        return h, S, tita   

    else:
        for l in range(alturas):
            L = str(l+1)
            # cuento el número de archivos en el directorio
            list = os.listdir(path+'/'+target+'/'+portadora+'/'+L)
            N_med = int(len(list))
            #print('# mediciones =', N_med)
            #print()

            # miro la longitud de cada mediciòn
            L_data = np.zeros(N_med, dtype=int) # vector en donde guardo la long de cada medición

            for j in range(N_med): 
                J = str(j+1)
                len_data = int(len(open(path+'/'+target+'/'+portadora+'/'+L+'/'+J+'.txt').readlines()))
                L_data[j] = len_data
                #print('medición ',target,' #'+J+' =',len_data)
            
            #-------------------------------------------------------------
            # Python program to check if all elements in a List are same 
            #-------------------------------------------------------------
            first = L_data[0]
            chk = True

            # Comparing each element with first item 
            for item in L_data:
                if first != item:
                    chk = False
                    break;
                    
            if (chk == True): 
                #print()
                #print("iguales, importo datos!!")
                S = np.zeros((alturas,N_med,int(len_data-1),11),dtype=float)

                for j0 in range(N_med):
                    J0 = str(j0+1)
                    S[l,j0,:,:] = np.genfromtxt(path+'/'+target+'/'+portadora+'/'+L+'/'+J+'.txt', delimiter=' ')

            else: 
                print("diferentes. Editar datos!!")      

        return h, S
  
```
