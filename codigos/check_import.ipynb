## Funcion que cuenta el numero de archivos en un directorio y si son todos iguales me importa los datos de la medicion

```python
import numpy as np
import os
```

```python
def check_import_ub(path, portadora, target, altura):
    '''
    Ubuntu - Funcion que chequea numero de mediciones y long. Importa datos
    Inputs:
    Path = path donde estan las mediciones
    (ej: /home/gordotony/Documents/UWB/tesis/mediciones_MM_DD)
    Portadora = Fc8 o Fc7
    altura = 1, 2, 3 o 4 segun altura de medicion
    target = S1, S2 o M2

    Output:
    S = devuelve el vector segun lo que importo
    '''

    # cuento el número de archivos en el directorio
    list = os.listdir(path+target+portadora+altura) # dir is your directory path
    N_med = int(len(list))
    print('# mediciones =', N_med)
    print()

    # miro la longitud de cada mediciòn
    L_data = np.zeros(N_med, dtype=int) # vector en donde guardo la long de cada medición

    for j in range(N_med): 
        J = str(j+1)
        len_data = int(len(open(path+target+portadora+altura+J+'.txt').readlines()))
        L_data[j] = len_data
        print('medición ',target,' #'+J+' =',len_data)
    
    S = np.zeros((N_med,int(len_data-1),11),dtype=float)
    
    # Python program to check if all elements in a List are same 
    first = L_data[0]
    chk = True

    # Comparing each element with first item 
    for item in L_data:
        if first != item:
            chk = False
            break;
              
    if (chk == True): 
        print("iguales, importo datos")

        for j0 in range(N_med):
            J0 = str(j0)
            S[j0,:,:] = np.genfromtxt(path+target+'/'+portadora+altura+target+'_'+J0+'.txt', delimiter=' ')

    else: 
        print("diferentes. Editar datos")      

    return S    
  
```
