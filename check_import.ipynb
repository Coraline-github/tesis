```python
def check_import(Path, target, portadora):
    
    '''
    Funcion que chequea numero de mediciones y long. Importa datos
    Inputs:
    Path = path donde estan las mediciones
    Portadora = Fc8 o Fc7
    target = (S1, S2, M2, X2) o (Z2, MZ2)

    Output:
    h = alturas de las mediciones
    S = devuelve el vector segun lo que import
    tita = angulo de incidencia (si es dif a 0)
    '''
    altura = os.path.join(Path,target,'h.txt')
    h = np.genfromtxt(altura, delimiter=' ')
    #print(h)

    if target == 'Z2' or target == 'MZ2':
        Tita = os.path.join(Path,target,'tita.txt')
        tita = np.genfromtxt(Tita, delimiter=' ')
        ang = len(tita)

        #suelo medir en una unica altura asi que no tengo en cuenta la variable h
        for l in range(ang):
            L = str(l+1)
            # cuento el número de archivos en el directorio
            paths = os.path.join(Path,target,portadora,'tita'+L)
            list = os.listdir(paths)
            N_med = int(len(list))
            #print('# mediciones =', N_med)
            #print()

            # miro la longitud de cada mediciòn
            Len_data = np.zeros(N_med, dtype=int) # vector en donde guardo la long de cada medición

            for j in range(N_med): 
                J = str(j+1)
                directorios = os.path.join(Path,target,portadora,'tita'+L,J+'.txt')
                len_data = int(len(open(directorios).readlines()))
                Len_data[j] = len_data
                #print('medición ',target,' #'+J+' =',len_data)
            
            print(Len_data)
            #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
            # Python program to check if all elements in a List are same 
            #-----------------------------------------------------------------
            first = Len_data[0]
            chk = True

            # Comparing each element with first item 
            for item in Len_data:
                if first != item:
                    chk = False
                    break;
                    
            if (chk == True): 
                print("iguales, importo datos!!")
                print()
            else: 
                print("diferentes. Editar datos!!")
            #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~

        S = np.zeros((ang, N_med, int(Len_data[0]-1),11),dtype=float)
        for l in range(ang):
            L=str(l+1)   
            for j in range(N_med):
                J=str(j+1)
                directorio = os.path.join(Path,target,portadora,'tita'+L,J+'.txt')
                S[l,j,:,:] =  np.genfromtxt(directorio, delimiter=' ') 

        return h, S, tita

    else:
        alturas = len(h)
        for l in range(alturas):
            L=str(l+1)
            # cuento el número de archivos en el directorio
            list = os.listdir(os.path.join(Path,target,portadora,L))
            N_med = int(len(list))
            print('altura =',h[l])
            print('# mediciones =', N_med)


            # miro la longitud de cada mediciòn
            Len_data = np.zeros(N_med, dtype=int) # vector en donde guardo la long de cada medición

            for j in range(N_med): 
                J = str(j+1)
                directorio = os.path.join(Path,target,portadora,L,J+'.txt')
                #print(directorio)
                len_data = int(len(open(directorio).readlines()))
                Len_data[j] = len_data
                #print('altura:',L,' medicion #'+J+' =',len_data)
            #print()
            print(Len_data)

            #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
            # Python program to check if all elements in a List are same 
            #-------------------------------------------------------------
            first = Len_data[0]
            chk = True

            # Comparing each element with first item 
            for item in Len_data:
                #print(item)
                if first != item: #!= : not equal
                    chk = False
                    break;
                    
            if (chk == True): 
                print("iguales, importo datos!!")
                print()
                
            else:
                print('editar datos!')
            #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
        
        S = np.zeros((alturas, N_med, int(Len_data[0]-1),11),dtype=float)
        for l in range(alturas):
            L=str(l+1)
            for j in range(N_med):
                J = str(j+1)
                directorio = os.path.join(Path,target,portadora,L,J+'.txt')
                S[l,j,:,:] = np.genfromtxt(directorio, delimiter=' ')

        return h, S     
```
